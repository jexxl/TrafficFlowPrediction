## ADF

```python
def diff(timeseries):
    timeseries_diff1 = timeseries.diff(1)
    timeseries_diff2 = timeseries_diff1.diff(1)

    timeseries_diff1 = timeseries_diff1.fillna(0)
    timeseries_diff2 = timeseries_diff2.fillna(0)

    timeseries_adf = sm.tsa.stattools.adfuller(timeseries.tolist())
    timeseries_diff1_adf = sm.tsa.stattools.adfuller(timeseries_diff1.tolist())
    timeseries_diff2_adf = sm.tsa.stattools.adfuller(timeseries_diff2.tolist())

    print('timeseries_adf : ', timeseries_adf)
    print('timeseries_diff1_adf : ', timeseries_diff1_adf)
    print('timeseries_diff2_adf : ', timeseries_diff2_adf)

diff(trend.dropna())  # 不需要差分
```

# model

## trend

```python
train_ts = ts.iloc[0:-test_size]
test_ts = ts.iloc[-test_size:]

history_lst = train_ts.tolist()
test_lst = test_ts.tolist()
trend_pred = []

order = auto_arima(trend.dropna()).order
print('order:', order)

_daily = sm.tsa.seasonal_decompose(history_lst, period=4*24, two_sided=False).seasonal
_no_daily = np.array(history_lst) - _daily
trend_arr = sm.tsa.seasonal_decompose(_no_daily, period=4*24*7, two_sided=False).trend
trend_arr = trend_arr[np.isfinite(trend_arr)]
model = sm.tsa.ARIMA(trend_arr, order=order).fit()
trend_pred.append(model.forecast()[0][0])

p = progressbar.ProgressBar()
for i in p(range(test_size-1)):
    history_lst.append(test_lst[i])
    _daily = sm.tsa.seasonal_decompose(history_lst, period=4*24, two_sided=False).seasonal
    _no_daily = np.array(history_lst) - _daily
    trend_arr = sm.tsa.seasonal_decompose(_no_daily, period=4*24*7, two_sided=False).trend
    trend_arr = trend_arr[np.isfinite(trend_arr)]
    model = sm.tsa.ARIMA(trend_arr, order=order).fit()
    trend_pred.append(model.forecast()[0][0])

trend_pred = np.array(trend_pred)
```

## residual

```python
train_ts = ts.iloc[0:-test_size]
test_ts = ts.iloc[-test_size:]

history_lst = train_ts.tolist()
test_lst = test_ts.tolist()
residual_pred = []

# order = auto_arima(residual.dropna()).order
order = (1,0,1)
print('order:', order)

_daily = sm.tsa.seasonal_decompose(history_lst, period=4*24, two_sided=False).seasonal
_no_daily = np.array(history_lst) - _daily
residual_arr = sm.tsa.seasonal_decompose(_no_daily, period=4*24*7, two_sided=False).resid
residual_arr = residual_arr[np.isfinite(residual_arr)]
model = sm.tsa.ARIMA(residual_arr, order=order).fit()
residual_pred.append(model.forecast()[0][0])

p = progressbar.ProgressBar()
for i in p(range(test_size-1)):
    history_lst.append(test_lst[i])
    _daily = sm.tsa.seasonal_decompose(history_lst, period=4*24, two_sided=False).seasonal
    _no_daily = np.array(history_lst) - _daily
    residual_arr = sm.tsa.seasonal_decompose(_no_daily, period=4*24*7, two_sided=False).resid
    residual_arr = residual_arr[np.isfinite(residual_arr)]
    model = sm.tsa.ARIMA(residual_arr, order=order).fit()
    residual_pred.append(model.forecast()[0][0])

residual_pred = np.array(residual_pred)
```

## daily

```python
train_ts = ts.iloc[0:-test_size]
test_ts = ts.iloc[-test_size:]

history_lst = train_ts.tolist()
test_lst = test_ts.tolist()
daily_pred = []

daily_arr = sm.tsa.seasonal_decompose(history_lst, period=4*24, two_sided=False).seasonal
daily_pred.append(daily_arr[-4*24])

p = progressbar.ProgressBar()
for i in p(range(test_size-1)):
    history_lst.append(test_lst[i])
    daily_arr = sm.tsa.seasonal_decompose(history_lst, period=4*24, two_sided=False).seasonal
    daily_pred.append(daily_arr[-4*24])

daily_pred = np.array(daily_pred)
```

## weekly

```python
train_ts = ts.iloc[0:-test_size]
test_ts = ts.iloc[-test_size:]

history_lst = train_ts.tolist()
test_lst = test_ts.tolist()
weekly_pred = []

_daily = sm.tsa.seasonal_decompose(history_lst, period=4*24, two_sided=False).seasonal
_no_daily = np.array(history_lst) - _daily
weekly_arr = sm.tsa.seasonal_decompose(_no_daily, period=4*24*7, two_sided=False).seasonal
weekly_pred.append(weekly_arr[-test_size])

p = progressbar.ProgressBar()
for i in p(range(test_size-1)):
    history_lst.append(test_lst[i])
    _daily = sm.tsa.seasonal_decompose(history_lst, period=4*24, two_sided=False).seasonal
    _no_daily = np.array(history_lst) - _daily
    weekly_arr = sm.tsa.seasonal_decompose(_no_daily, period=4*24*7, two_sided=False).seasonal
    weekly_pred.append(weekly_arr[-test_size])
weekly_pred = np.array(weekly_pred)
```

## conclusion

有两个问题：

- weekly拟合得不好

- residual使用arima不易拟合

首先，区分daily与weekly是为了更好地刻画交通流量的周期特性，然而在滚动预测的过程中，由于周数较少，每增加一个观测值，weekly都有一些变化，难以用上一周的值来预测。

其次，arima在上一个测试中对residual的拟合效果不佳，在这里也是一样。并且在代码运行的过程中频繁报错，只好将order降到了(1,0,1)。

**因此，mstl是完全不需要的，将交通流量拆分为trend, daily, residual已经足够了。接下来需要对residual进行更加准确的预测。**
