# 所有路段（多线程）

```python
def test_arima(idx, test_len=4*24*7):
    
    ts = get_flow_series_by_idx(idx)
    ts = ts.resample('15min').sum() # 15min粒度
    
    arima_order = auto_arima(ts,).order  # 自动遍历arima参数
    print('#', idx, arima_order)
    
    train_ts = ts.iloc[0:-test_len]
    test_ts = ts.iloc[-test_len:]
    
    history_lst = train_ts.tolist()
    test_lst = test_ts.tolist()
    y_pred = []

    model = ARIMA(history_lst, order=arima_order)
    result_ARIMA = model.fit(disp=-1)
    y_pred.append(result_ARIMA.forecast()[0][0])

    for i in range(len(test_lst) - 1):
        print('#',idx,'at',i)
        history_lst.append(test_lst[i])
        model = ARIMA(history_lst, order=arima_order)
        result_ARIMA = model.fit(disp=-1)
        y_pred.append(result_ARIMA.forecast()[0][0])
    
    y_true = np.array(test_lst)
    y_pred = np.array(y_pred)
    
    return {
        'mae': mae(y_true, y_pred),
        'rmse': rmse(y_true, y_pred),
        'mape': mape(y_true, y_pred),
    }
```

```python
from IPython.lib import backgroundjobs as bg
jobs = bg.BackgroundJobManager()

arima_result = pd.DataFrame(columns=['mae','rmse','mape'])

def run_test_arima(idx):
    arima_result.loc[idx] = test_arima(idx)
    
for idx in range(10):
    jobs.new(run_test_arima, idx)
```
